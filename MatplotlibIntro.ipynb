```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.plot(x,y, label='2X', color = 'red', linewidth=2, linestyle= '--',marker ='.', markersize=10, markeredgecolor='blue')
plt.title('Our First Graph!', fontdict = {'fontname':'Comic Sans MS','fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
#plt.plot(x,y, label='2X', color = 'red', linewidth=2, linestyle= '--',marker ='.', markersize=10, markeredgecolor='blue')

plt.plot(x,y,'b^--',label='2X')

##line number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, 'r', label = 'X^2')

plt.title('Our First Graph!', fontdict = {'fontname':'Comic Sans MS','fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
#plt.plot(x,y, label='2X', color = 'red', linewidth=2, linestyle= '--',marker ='.', markersize=10, markeredgecolor='blue')

plt.plot(x,y,'b^--',label='2X')

##line number 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2[:6], x2[:6]**2, 'r', label ='X^2')
plt.plot(x2[5:], x2[5:]**2, 'r--')

plt.title('Our First Graph!', fontdict = {'fontname':'Comic Sans MS','fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4,])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.show()
```

## Bar Chart


```python
labels = ['A', 'B', 'C']
values = [1,4,2]
plt.bar(labels, values)
plt.figure(figsize=(6,4))
```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar(labels, values)

patterns = ['/','O','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(6,4))

plt.show()
```
